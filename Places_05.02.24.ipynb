```python
import pandas as pd
df = pd.read_csv('places.csv')
df
```

```python
newdf = df.drop(columns = ['Unnamed: 0','Zone','State','City','Name','Establishment Year','time needed to visit in hrs','Airport with 50km Radius','Weekly Off','Significance','DSLR Allowed','Best Time to visit'])
newdf
```

```python
newdf.nunique()
```

```python
df1 = newdf.drop(newdf[newdf['Entrance Fee in INR'] == 0].index)
df2 = df1.drop(columns = ['Entrance Fee in INR'])
df1
df2
```

```python
df3 = pd.get_dummies(df1,dtype=int)
df3
```

```python
y = df1['Entrance Fee in INR']
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df3,y, test_size=0.33)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
reg2 = Lasso(alpha = 25,max_iter = 100, tol = 0.1)
reg2.fit(x_train,y_train)
```

```python
reg2.score(x_test,y_test)
```

```python
reg2.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge
reg3 = Ridge(alpha = 25,max_iter = 100, tol = 0.1)
reg3.fit(x_train,y_train)
```

```python
reg3.score(x_test,y_test)
```

```python
reg3.score(x_train,y_train)
```
